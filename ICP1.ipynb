# **Data Cleaning**

```
data.info()
```

As we can see there are no null values in the data, except for the Unnamed column. It is not much useful for data analysis, hence we will drop it.

```
data.drop(columns=['Unnamed: 32'], inplace=True)
```

```
data.info()
```

# **Descriptive Statistical Analysis - Mean, Median and Mode**

```
mean = data['perimeter_se'].mean()
median = data['perimeter_se'].median()
mode = data['perimeter_se'].mode()
```

```
print("Perimeter se mean= ",mean)
print("Perimeter se median= ", median)
print("Perimeter se mode= ", mode)
```
